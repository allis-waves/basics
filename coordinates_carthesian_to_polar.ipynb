```python
import matplotlib.pyplot as plt
import numpy as np

# add functions to #1 functions#, change # SETTINGS # if necessary

#############
# FUNCTIONS #
#############

functions = [
    lambda x: x * 2,                    # y = x * 2
    lambda x: x**2,                     # y = x^2
    lambda x: 20 * np.cos(x*2) + 60,    # y = [amplitude] * cos( x * [periods] )
]

###################################
# POLAR COORDINATE TRANSFORMATION #
###################################

def cartesian_to_polar(x, y):
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    return theta, r

############
# SETTINGS #
############

x = np.linspace(-10, 10, 400)   # range of x values 
theta_values = np.linspace(0, 2*np.pi, 400) # range of theta values for polar plot
fig, (ax_cart, ax_polar) = plt.subplots(1, 2, figsize=(12, 6))  # figure size


############
# PLOTTING #
############

# cartesian
for func in functions:
    y = func(x)
    ax_cart.plot(x, y, label=f'func {functions.index(func)}')
ax_cart.set_title('Cartesian Coordinate System')
ax_cart.axhline(0, color='black',linewidth=0.5)
ax_cart.axvline(0, color='black',linewidth=0.5)
ax_cart.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax_cart.legend()

# polar
ax_polar = plt.subplot(122, polar=True)
for func in functions:
    x_for_polar = np.linspace(-10, 10, 400)  # Use x range to generate r values
    y_for_polar = func(x_for_polar)
    theta_for_polar, r_for_polar = cartesian_to_polar(x_for_polar, y_for_polar)
    ax_polar.plot(theta_values, r_for_polar, label=f'func {functions.index(func)}')  # Plot using theta_values for the circle
ax_polar.set_title('Polar Coordinate System')
ax_polar.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax_polar.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax_polar.legend()

plt.show()
```

```python
# theta variation with x

import matplotlib.pyplot as plt
import numpy as np

# Define the functions list as before
functions = [
    lambda x: x,            # y = x
    lambda x: x**2,         # y = x^2
    lambda x: np.cos(x),    # y = cos(x)
]

# Cartesian to Polar conversion function as before
def cartesian_to_polar(x, y):
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    return theta, r

# Plotting
x = np.linspace(-10, 10, 400)
fig, axs = plt.subplots(1, 3, figsize=(18, 6))  # Adjusted for an additional plot

# Cartesian Plot
ax_cart = axs[0]
for func in functions:
    y = func(x)
    ax_cart.plot(x, y, label=f'func {functions.index(func)}')
ax_cart.set_title('Cartesian Coordinate System')
ax_cart.axhline(0, color='black',linewidth=0.5)
ax_cart.axvline(0, color='black',linewidth=0.5)
ax_cart.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax_cart.legend()

# Polar Plot
ax_polar = axs[1] = plt.subplot(132, polar=True)
for func in functions:
    y = func(x)
    theta, r = cartesian_to_polar(x, y)
    ax_polar.plot(theta, r, label=f'func {functions.index(func)}')
ax_polar.set_title('Polar Coordinate System')
ax_polar.set_rlabel_position(-22.5)
ax_polar.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax_polar.legend()

# Theta Plot (as an example of how to visualize theta variation)
ax_theta = axs[2]
for func in functions:
    y = func(x)
    theta, r = cartesian_to_polar(x, y)
    ax_theta.plot(x, theta, label=f'func {functions.index(func)} Theta')
ax_theta.set_title('Theta Variation with x')
ax_theta.axhline(0, color='black',linewidth=0.5)
ax_theta.axvline(0, color='black',linewidth=0.5)
ax_theta.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax_theta.legend()

plt.tight_layout()
plt.show()
```

```python
# automatic plotting as polar projection matplotlib

import matplotlib.pyplot as plt
import numpy as np

# Define a linear function for demonstration
def linear_function(x):
    return 2 * x + 1

# Generate x values
x_values = np.linspace(0, 10, 100)

# Calculate corresponding y values from the linear function
y_values = linear_function(x_values)

# Transform y_values to fit within the range of 0 to 2 for mapping onto a circle
max_y = np.max(y_values)
theta_values = 2 * np.pi * (y_values / max_y)

# Plotting
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot each (radius, angle) pair
for r, theta in zip(x_values, theta_values):
    ax.plot([theta], [r], 'ro')  # 'ro' plots red dots

# Set the title
ax.set_title('Transformed Line in Polar Coordinates', va='bottom')

plt.show()
```
